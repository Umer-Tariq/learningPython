MinMax Algorithm

```python
import math

class node:
    def __init__(self, value = None):
        self.value = value
        self.children = []
        self.minimax_value = None
def minMax(node, depth, maximizing = True):
    if depth == 0 or len(node.children) == 0:
        return node.value
    child_value = []
    if maximizing:
        child_value.clear()
        for child in node.children:
            child_value.append(minMax(child, depth - 1, False))
        node.minimax_value = max(child_value)
        return node.minimax_value
    else:
        child_value.clear()
        for child in node.children:
            child_value.append(minMax(child, depth - 1, True))
        node.minimax_value = min(child_value)
        return node.minimax_value
        
root = node('A')
n1 = node('B')
n2 = node('C')
root.children = [n1, n2]

n3 = node('D')
n4 = node('E')
n1.children = [n3, n4]

n5 = node('F')
n6 = node('G')
n2.children = [n5, n6]


n7 = node(2)
n8 = node(3)
n9 = node(5)
n10 = node(9)
n3.children = [n7, n8]
n4.children = [n9, n10]


n11 = node(0)
n12 = node(4)
n13 = node(7)
n14 = node(5)
n5.children = [n11, n12]
n6.children = [n13, n14]

minMax(root, 3)
print("Minimax Values:")
print("A: ", root.minimax_value)
print("B: " ,n1.minimax_value)
print("C: " ,n2.minimax_value)
print("D: " ,n3.minimax_value)
print("E: " ,n4.minimax_value)
print("F: " ,n5.minimax_value)
print("G: " ,n6.minimax_value)
```

```python
import math

class node:
    def __init__(self, value):
        self.value = value
        self.children = None
        self.minimax_val = None
def alpha_beta(node, depth, alpha, beta, maximizing = True):
    if depth == 0 or len(node.children) == 0:
        return node.value
    if maximizing:
        value = -math.inf
        for child in node.children:
            value = max(value, alpha_beta(child, depth - 1, alpha, beta, False))
            alpha = max(alpha, value)
            if beta <= alpha:
                print('Pruned node: ', child.value)
                break
        node.minimax_val = value
        return value
    else:
            value = math.inf
            for child in node.children:
                value = min(value, alpha_beta(child, depth - 1, alpha, beta, True))
                beta = min(beta, value)
                if beta < alpha:
                    print('Pruned node: ', child.value)
                    break
            node.minimax_val = value
            return value
        
root = node('A')
n1 = node('B')
n2 = node('C')
root.children = [n1, n2]

n3 = node('D')
n4 = node('E')
n1.children = [n3, n4]

n5 = node('F')
n6 = node('G')
n2.children = [n5, n6]


n7 = node(2)
n8 = node(3)
n9 = node(5)
n10 = node(9)
n3.children = [n7, n8]
n4.children = [n9, n10]


n11 = node(0)
n12 = node(4)
n13 = node(7)
n14 = node(5)
n5.children = [n11, n12]
n6.children = [n13, n14]

alpha_beta(root, 3, -math.inf, math.inf)
print("Minimax Values:")
print("A: ", root.minimax_val)
print("B: " ,n1.minimax_val)
print("C: " ,n2.minimax_val)
print("D: " ,n3.minimax_val)
print("E: " ,n4.minimax_val)
print("F: " ,n5.minimax_val)
print("G: " ,n6.minimax_val)
```
