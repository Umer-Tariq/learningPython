```python
import string
import random
```

```python
def getString(n):
    arr = string.ascii_letters + string.digits + string.punctuation
    stri = []
    for i in range(n):
        stri.append(random.choice(arr))
    return stri

def generate_population(n, pop_size):
    population = []
    for i in range(pop_size):
        chromosome = getString(n)
        population.append(chromosome)
    return population

def get_score(chromosome, target):
    count = 0
    for i in range(len(target)):
        if target[i] != chromosome[i]:
            count += 1
    return count

def fitness(population, target):
    fitness_score = []
    for chromosome in population:
        score = get_score(chromosome, target)
        if score == 0:
            print('Target achieved!!!')
            return True
        fitness_score.append((score, chromosome))
    fitness_score.sort()
    return fitness_score
        
def mate(fitness_score):
    index1 = 4
    index2 = 7
    substr1 = fitness_score[1][1][index1:index2]
    for parent2 in fitness_score:
        substr2 = parent2[1][index1:index2]
        parent2[1][index1:index2] = substr1
        fitness_score[1][1][index1:index2] = substr2
    
def mutate(fitness_score):
    arr = string.ascii_letters + string.digits + string.punctuation
    index = random.randint(0, len(fitness_score[0][1]) - 1)
    for item in fitness_score:
        c = random.choice(arr)
        item[1][index] = c
def extractPopulation(fitness_score):
    population = []
    for item in fitness_score:
        population.append(item[1])
    return population

target = 'Artificial Intelligence Lab'
pop_size = 70
itr = 100
population = generate_population(len(target), pop_size)
i = 0
while i < itr:
    fitness_score = fitness(population, target)
    if fitness_score == True:
        break
    mate(fitness_score)
    mutate(fitness_score)
    i += 1
    population = extractPopulation(fitness_score)
    
```

```python
def getChromosome(n):
    arr = [0, 1]
    chromosome = []
    for i in range(n):
        chromosome.append(random.choice(arr))
    return chromosome

def generate_population(n, pop_size):
    population = []
    for i in range(pop_size):
        population.append(getChromosome(n))
    return population
    
def cross_over(fitness_score):
    index = 1
    for parent2 in fitness_score:
        substr1 = fitness_score[0][1][index:]
        substr2 = parent2[1][index:]
        parent2[1][index:] = substr1
        fitness_score[0][1][index:] = substr2
    
def fitness(population):
    fitness_score = []
    for item in population:
        score = 16*item[0] + 8*item[1] + 4*item[2] + 2*item[3] + 1*item[4]
        fitness_score.append((score, item))
    fitness_score.sort()
    return fitness_score
    
def mutate(fitness_score):
    arr = [0, 1]
    for item in fitness_score:
        index1 = 2
        index2 = 4 
        while index1 <= index2:
            item[1][index1] = random.choice(arr)
            index1 += 1
def getPopulation(fitness_score):
    population = []
    for item in fitness_score:
        population.append(item[1])
    return population
    
n = 5
pop_size = 6
population = generate_population(n ,pop_size)
i = 0
itr = 100
while i < itr:
    fitness_score = fitness(population)
    cross_over(fitness_score)
    mutate(fitness_score)
    population = getPopulation(fitness_score)
    i += 1
```
