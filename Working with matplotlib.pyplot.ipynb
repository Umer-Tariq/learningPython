```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([1, 2, 3, 4])
plt.plot(x, y, color='green', marker='o',linewidth=2, markersize=12)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, marker = 'x')
x2 = np.array([5, 6, 7, 8])
y2 = np.array([5, 6, 7, 8])
plt.subplot(1, 2, 2)
plt.plot(x2, y2, marker = 'o', color = 'green')
plt.show()
```

```python
#showing only markers
plt.plot(x, y, marker = 'x', color = 'green', linestyle = 'dotted')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Matplotlib Demo')
plt.grid()
plt.show()
```

```python
#string format
plt.plot(x, y, 'x--r')
```

```python
#scatter plot 
x = np.arange(0, 10, 2)
x2 = np.arange(1, 11, 2)
plt.scatter(x, x2)
x = np.arange(0, 20, 3)
x2 = np.arange(1, 21, 3)
plt.scatter(x, x2)
plt.show()
```

```python
#bar chart
x = np.array(["A", "B", "C", "D"])
values = np.array([10, 7, 9, 2])
plt.bar(x, values, color = 'red', width = 0.5)
plt.show()
plt.barh(x, values, color = 'green', height = 0.1)
plt.show()
```

```python
#histograms
x = np.random.normal(170, 10, 250)
plt.hist(x, bins = 100, rwidth = 1.5)
plt.title('Histogram example')
plt.xlabel('Age Range')
plt.ylabel('Height')
plt.show()
```

```python
#Pie chart
percentages = [20, 30, 40, 10]
mylabels = ['Projects', 'Gym', 'Skill development', 'Uni work']
angle = 90
myexplode = [0.2, 0.1, 0.2, 0.1]
plt.pie(percentages, labels = mylabels, startangle = angle, explode = myexplode)
plt.legend(loc = 'right')
plt.show()
```
