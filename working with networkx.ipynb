Implementing Networkx library

```python
import networkx as nx
```

```python
DG = nx.DiGraph()
DG.add_nodes_from([0, 1, 2, 3, 4])
DG.add_edges_from([(0, 1), (0, 2), (1, 3), (1, 4)])
nx.draw(DG, with_labels = True, node_color = 'lightblue', font_weight = 'bold')
```

```python
G = nx.Graph()
G.add_nodes_from([0, 1, 2, 3, 4])
G.add_edges_from([(0, 1), (1, 3), (0, 2), (1, 4)])
nx.draw(G, with_labels = True, node_color = 'lightblue', font_weight = 'bold')
```

```python
edges = nx.bfs_edges(G, 0)
nodes = []
leaves = []
for edge in edges:
    print(edge)
    if edge[1] not in nodes:
        nodes.append(edge[1])
    if edge[0] not in leaves:
        leaves.append(edge[0])
nodes.extend(leaves)
print(set(nodes))
```
